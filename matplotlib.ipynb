```python
import matplotlib.pyplot as plt
import numpy as np

t1 = np.linspace(-np.pi, np.pi, 100)
# t2 = np.arange(-np.pi, np.pi, 0.1)

noise = [np.random.rand() / 10 for t in t1]

y1 = np.sin(t1)
y2 = np.cos(t1)
y3 = np.sin(t1) + np.cos(t1) + noise

plt.plot(t1, y1, "r")
plt.plot(t1, y2, "b")
plt.plot(t1, y3, "g")
```
