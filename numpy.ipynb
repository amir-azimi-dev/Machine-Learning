```python
import numpy as np

a = np.arange(12)

a = a.reshape(3, 4)
print(a, "\n")
print(a.shape, "\n")
print(a.ndim, "\n")
print(a.size, "\n")
print(a.dtype, "\n")
print(a > 9, "\n")
```

```python
import numpy as np

a = np.zeros((3, 2))
b = np.ones((3, 2))
c = np.full((3, 2), 7)
print(a, "\n")
print(b, "\n")
print(c, "\n")
```

```python
import numpy as np

a = np.array([10, 20, 30, 40]).reshape(2, 2)
b = np.arange(4).reshape(2, 2)

print(a, "\n")
print(b, "\n\n")
print(a + b, "\n")
print(a - b, "\n")
print(a * b, "\n")
print(a @ b, "\n")
print(b / a)
```
